```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2*np.pi)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.xlabel("x value")
plt.ylabel("y value")

plt.title("sin/cos")

plt.plot(x,y_sin, label = "sin")
plt.plot(x,y_cos, label = "cos",linestyle="dashed")
plt.legend()
plt.show()
```
